```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

datasets = df_norm['Dataset'].unique()
epsilons = sorted(df_norm['Epsilon'].unique())
dfs = {'Normal': df_norm, 'Adv. Trained': df_pgd, 'GAN+Adv': df_gan}
colors = ['skyblue', 'salmon', 'lightgreen']

for dataset in datasets:
    df_list = []
    for name, df in dfs.items():
        temp = df[df['Dataset'] == dataset].copy()
        temp['Method'] = name
        df_list.append(temp)
    df_combined = pd.concat(df_list)

    n_eps = len(epsilons)
    n_cols = 2
    n_rows = math.ceil(n_eps / n_cols)

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(6*n_cols, 5*n_rows), sharey=True)
    axes = axes.flatten()  # flatten to simplify indexing

    for j, eps in enumerate(epsilons):
        ax = axes[j]
        df_eps = df_combined[df_combined['Epsilon'] == eps]
        sns.barplot(
            data=df_eps,
            x='Attack',
            y='F1',
            hue='Method',
            palette=colors,
            ax=ax
        )
        ax.set_title(f'Epsilon = {eps}')
        ax.set_ylim(0, 1)
        ax.set_ylabel('F1 Score')
        ax.set_xlabel('Attack Type')
        if j == n_eps - 1:
            ax.legend(title='Method')
        else:
            ax.get_legend().remove()

    # Remove empty subplots if epsilons are odd
    for k in range(n_eps, n_rows*n_cols):
        fig.delaxes(axes[k])

    plt.suptitle(f'F1 Score Comparison for Dataset: {dataset}', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```
