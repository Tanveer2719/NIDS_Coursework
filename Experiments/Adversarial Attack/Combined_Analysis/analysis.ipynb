```python
import matplotlib.pyplot as plt
import pandas as pd

# Models and attacks
models = ["Adv Training", "GAN+Adv Training"]
attacks = ["FGSM", "PGD", "MI-FGSM"]

# Function to calculate percentage improvement
def calculate_improvement_percent(dfs, metric="F1"):
    improvements = {}
    for attack in attacks:
        improvements[attack] = {}
        # baseline (normal training)
        base = dfs[0][dfs[0]["Attack"] == attack].set_index("Epsilon")[metric]
        for model_df, model_name in zip(dfs[1:], models):  # skip normal
            model_scores = model_df[model_df["Attack"] == attack].set_index("Epsilon")[metric]
            improvement_percent = ((model_scores - base) / base) * 100
            improvements[attack][model_name] = improvement_percent
    return improvements

# Plot percentage improvement
def plot_improvement_percent(improvements, dataset_name):
    fig, axes = plt.subplots(1, 3, figsize=(18, 7), sharey=True)
    
    for ax, attack in zip(axes, attacks):
        df = pd.DataFrame(improvements[attack])
        df.plot(kind="bar", ax=ax, width=0.7)
        ax.set_title(f"{attack}", fontsize=13)
        ax.set_xlabel("Epsilon (ε)", fontsize=12)
        ax.set_ylabel("F1 Improvement (%)" if ax==axes[0] else "", fontsize=12)
        ax.grid(True, linestyle="--", alpha=0.6)
    
    fig.suptitle(f"{dataset_name} - Performance improvement with respect to  Normal Training", fontsize=16)
    axes[2].legend(fontsize=11, loc="upper left")
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

# ---- UNSW ----
improvements_unsw_percent = calculate_improvement_percent(dataframes_unsw)
plot_improvement_percent(improvements_unsw_percent, "UNSW-NB15")

# ---- NSL ----
improvements_nsl_percent = calculate_improvement_percent(dataframes_nsl)
plot_improvement_percent(improvements_nsl_percent, "NSL-KDD")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Models and attacks
models = ["Normal Training", "PGD Training", "GAN Training"]
attacks = ["FGSM", "PGD", "MI-FGSM"]

# Clean data F1 scores
clean_f1 = {
    "UNSW-NB15": 0.948,
    "NSL-KDD": 0.9911
}

# Markers and linestyles
model_styles = {
    "Normal Training": {"marker": "o", "linestyle": "--"},
    "PGD Training": {"marker": "^", "linestyle": "-."},
    "GAN Training": {"marker": "s", "linestyle": ":"}
}

# Function to calculate degradation (%)
def calculate_degradation_percent(dfs, dataset_name, metric="F1"):
    degradation = {}
    for attack in attacks:
        degradation[attack] = {}
        for df, model_name in zip(dfs, models):
            model_scores = df[df["Attack"] == attack].set_index("Epsilon")[metric]
            degradation_percent = ((clean_f1[dataset_name] - model_scores) / clean_f1[dataset_name]) * 100
            degradation[attack][model_name] = degradation_percent
    return degradation

# Plot line plot for degradation
def plot_degradation_line(degradation, dataset_name):
    fig, axes = plt.subplots(1, 3, figsize=(18, 7), sharey=True)
    
    for ax, attack in zip(axes, attacks):
        # Plot each model line
        for model_name, style in model_styles.items():
            epsilons = degradation[attack][model_name].index
            values = degradation[attack][model_name].values
            ax.plot(
                epsilons, values,
                marker=style["marker"],
                linestyle=style["linestyle"],
                markersize=12,            # larger marker
                markerfacecolor='none',   # hollow marker
                markeredgewidth=2,        # thicker edge
                linewidth=2,
                label=model_name
            )
            
            # Annotate last point
            ax.text(epsilons[-1], values[-1]+0.5, f"{values[-1]:.2f}%", fontsize=10,
                    ha="center", va="bottom")

        # Plot horizontal line for clean F1
        ax.axhline(0, color="black", linestyle="--", linewidth=1, alpha=0.5)

        ax.set_title(f"{attack}", fontsize=13)
        ax.set_xlabel("Epsilon (ε)", fontsize=12)
        if ax==axes[0]:
            ax.set_ylabel("F1 Degradation (%)", fontsize=12)
        ax.grid(True, linestyle="--", alpha=0.6)
        ax.legend(fontsize=11)
    
    fig.suptitle(f"{dataset_name} - F1 Degradation relative to Clean Data (Clean F1 = {clean_f1[dataset_name]:.3f})",
                 fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

# ---- UNSW ----
degradation_unsw = calculate_degradation_percent(dataframes_unsw, "UNSW-NB15")
plot_degradation_line(degradation_unsw, "UNSW-NB15")

# ---- NSL ----
degradation_nsl = calculate_degradation_percent(dataframes_nsl, "NSL-KDD")
plot_degradation_line(degradation_nsl, "NSL-KDD")
```
